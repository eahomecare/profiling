```python
import pandas as pd
from IPython.display import display, HTML


pd.set_option('max_colwidth', 10)

file_path = 'booking_new.csv'
df = pd.read_csv(file_path)

display(HTML(df.tail().to_html()))

```

```python
print(df.info())
```

```python
df['DurationSlot'] = pd.to_numeric(df['durationSlot'], errors='coerce')
grouped_data = df.groupby('ServiceType').agg({'orderId': 'count', 'durationSlot': 'mean'})

grouped_data
```

```python
status_counts = df.groupby('Status')['OrderId'].count()
print(status_counts)
```

```python
service_status_counts = df.groupby(['ServiceType', 'Status'])['OrderId'].count()
print(service_status_counts)
```

```python
call_ticket_counts = df.groupby(['callTicketType', 'callTicketSubType'])['OrderId'].count()
print(call_ticket_counts)
```

```python
customer_ticket_counts = df.groupby('CustomerName')['OrderId'].count().sort_values(ascending=False)
print(customer_ticket_counts)
```

```python
service_counts = df['ServiceType'].value_counts()
print(service_counts)
```

```python
source_counts = df['SourceOfBooking'].value_counts()
print(source_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df['serviceDate'] = pd.to_datetime(df['ServiceDate'], errors='coerce')

# Drop rows with NaN values in 'serviceDate' if needed
# df = df.dropna(subset=['serviceDate'])

bookings_over_time = df.groupby('ServiceDate')['OrderId'].count()

plt.figure(figsize=(12, 6))
bookings_over_time.plot(kind='line')
plt.title('Booking Frequency Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
import matplotlib.pyplot as plt


customer_booking_counts = df.groupby('CustomerName')['OrderId'].count().sort_values(ascending=False)

top_100_customers = customer_booking_counts.head(100)

plt.figure(figsize=(24, 6))
top_100_customers.plot(kind='bar')
plt.title('Number of Bookings for Top 100 Customers')
plt.xlabel('Customer')
plt.ylabel('Number of Bookings')
plt.show()

```

```python
repeat_customers_over_time = df.groupby(['CustomerName', 'ServiceDate'])['OrderId'].count()
repeat_customers_over_time = repeat_customers_over_time[repeat_customers_over_time > 1].groupby('CustomerName').count()



plt.figure(figsize=(24, 6))
repeat_customers_over_time.plot(kind='line')
plt.title('Number of Repeat Bookings Over Time')
plt.xlabel('Customer')
plt.ylabel('Number of Repeat Bookings')
plt.show()
```

```python
service_type_preferences = df['ServiceType'].value_counts()

plt.figure(figsize=(24, 6))
service_type_preferences.plot(kind='bar')
plt.title('Service Type Preferences')
plt.xlabel('Service Type')
plt.ylabel('Number of Bookings')
plt.show()
```
